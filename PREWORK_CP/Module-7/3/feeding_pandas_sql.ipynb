```python
# Initial imports
import pandas as pd
import hvplot.pandas
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os
```

## Create a connection to the `university` database

```python
# Load .env enviroment variables and get SQL coonection
load_dotenv('../../.env')
db_url = os.getenv("db_url")
```

```python
# Define the databaser URL
# Create the engine object
engine = create_engine(db_url)
```

## Create a DataFrame with the names of all students

```python
# Write the SQL query
sql_query = "SELECT count(region_id) as NoRegions, ag.first_name " \
            "FROM agent_region_junction a " \
            "JOIN agents ag on ag.agent_id = a.agent_id " \
            "GROUP BY ag.first_name"

# Read the SQL query into a DataFrame
agents_df = pd.read_sql(sql_query, engine)

# Show the DataFrame's head

agents_df
```

```python
# Create the bar chart usig hvplot

agents_df.hvplot.bar(
    x = 'first_name',
    y='noregions',
    xlabel='agent first name',
    ylabel='count regions',
    title = 'Regions per Agents',
    color = 'noregions',
)
```
