# Visualizing Rolling Statistics with Pandas DataFrames


## 1. Import Libraries and Dependencies

```python
# Import libraries and dependencies
import pandas as pd
import numpy as np
from pathlib import Path
%matplotlib inline
```

## 2. Create a Path to the File Using Pathlib.

```python
# Read the MSFT data, set the `date` as the index
msft_path = Path('../Resources/MSFT.csv')
```

## 3. Read the CSV into a Pandas DataFrame.

```python
# Read the MSFT CSV into a DataFrame while formatting and setting the `date` column as the index
msft = pd.read_csv(msft_path, header=0, index_col="date", infer_datetime_format=True, parse_dates=True)
```

## 4. Drop Extra Columns

```python
# Use the `drop` function to drop the extra columns
delete_columns = ['volume', 'open', 'high', 'low']
for column in delete_columns:
    msft.drop(column, axis=1, inplace=True)
print(msft)
```

## 5. Plot Daily Closing Price of MSFT

```python
# Plot the daily closing prices of MSFT
msft.plot()
```

## 6. Rolling 7-Day Mean of MSFT Closing Prices

```python
# Plot a rolling 7-day mean of MSFT's closing prices
msft.rolling(window=7).mean().plot()
```

## 7. Rolling 7-Day Standard Deviation of MSFT Closing Prices

```python
# Plot a rolling 7-day std dev of MSFT's closing prices
msft.rolling(window=7).std().plot()
```

## 8. Rolling 30-Day Mean and Standard Deviation of MSFT Closing Prices

```python
# Plot a rolling 30-day mean of MSFT's closing prices
msft.rolling(window=30).mean().plot()
```

## 9. Rolling 30-Day Standard Deviation of MSFT Closing Prices

```python
# Plot a rolling 30-day std dev of MSFT's closing prices
msft.rolling(window=30).std().plot()
```

## 10. Rolling 180-Day Mean of TSLA Closing Prices

```python
# Plot a rolling 180-day mean of MSFT's closing prices
msft.rolling(window=180).mean().plot()
```

## 11. Rolling 180-Day Standard Deviation of TSLA Closing Prices

```python
# Plot a rolling 180-day std dev of MSFTs closing prices
msft.rolling(window=180).std().plot()
```

## 12. Overlay Daily Closing Price and 180-Day Rolling Mean of MSFT

```python
# Set figure of the daily closing prices of MSFT
ax = msft.plot()

# Plot 180-Day Rolling Mean on the same figure
msft.rolling(window=180).mean().plot(ax=ax)

# Set the legend of the figure
ax.legend(["msft", "msft 180 Day Mean"]);
```
