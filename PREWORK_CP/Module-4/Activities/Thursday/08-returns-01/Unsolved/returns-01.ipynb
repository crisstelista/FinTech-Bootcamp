# Calculating Stock Returns using Pandas DataFrames

## 1. Import Libaries and Dependencies

```python
# Import necessary libraries and dependencies
import pandas as pd
import numpy as np
from pathlib import Path
%matplotlib inline
```

## 2. Create a Path to the File Using Pathlib

```python
# Use the Pathlib libary to set the path to the CSV
csv_path = Path('../Resources/MSFT.csv')
```

## 3. Read CSV in as DataFrame

```python
# Read in the CSV as a DataFrame
csv_file = pd.read_csv(csv_path, header=0, index_col="date", infer_datetime_format=True, parse_dates=True)
```

## 4. Drop Extra Columns

```python
# Drop unnecessary columns
delete_columns = ['volume', 'open', 'high', 'low']
for column in delete_columns:
    csv_file.drop(column, axis=1, inplace=True)
print(csv_file)
```

## 5. Set Date as Index

```python
# Set the date as the index to the DataFrame
csv_file.sort_index()
print(csv_file)
```

```python
# Drop the extra date column
```

## 7. Plot Daily Close

```python
# Plot the daily closing stock price of the S&P 500 over the last 5 years
csv_file['close'].plot()
```

## 8. Calculate Daily Returns

```python
# Calculate the daily return using the 'pct_change()' function
daily_returns = csv_file.pct_change().dropna()
daily_returns.head()
```

## 9. Plot Daily Returns

```python
# Plot the daily returns of MSFT over the last 10 years
daily_returns.plot()
```

## 10. Calculate Cumulative Returns

```python
# Calculate the cumulative returns using the 'cumprod()' function
cum_returns = (1+daily_returns).cumprod()
cum_returns.head()
```

## 11. Plot Cumulative Returns

```python
# Plot the daily returns of the S&P 500 over the last 5 years
cum_returns.plot()
```
